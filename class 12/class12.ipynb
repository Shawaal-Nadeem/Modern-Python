# File Handling

```python
from typing import TextIO
data:TextIO=open('./abc.txt')
print(data.read())

data.close()
```

```python
# If we have a large data and we don't remember to close file {data.close()} then we use with block approach, it automatically close file.

with open('./abc.txt','r') as file:  # r means read and if we put othing then it also same means
    print(file.readlines())
```

```python

with open('./abc1.txt','w') as file:  # w means write and overwrite file
    file.write("Pakistan Zindabaad")
```

```python
with open('./abc1.txt','r+') as file:
    print(file.read())
    file.write("Today is Iqbal Day")
    file.seek(0) # start cursor from 0 index
    print(file.read())
```

```python
with open('./abc1.txt','a') as file:
    file.write("Hello")
```
